```python
import numpy as np
import seaborn as sns; sns.set()
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [12, 10]

font = {'size'   : 15, }
axes = {'labelsize': 'large', 'titlesize': 'large'}

mpl.rc('font', **font)
mpl.rc('axes', **axes)

import pandas as pd
import data_utils
import glob
import os
import matplotlib.style as style
```

```python
wd = "/home/behzad/Documents/barnes_lab/cplusplus_software/speed_test/repressilator/cpp/"
data_dir = wd + "output/two_species_stable_4/Population_0/"
states_dir = data_dir + "simulation_states/"

output_dir = wd + "data_analysis_notebook/figures/figure1_parts/"
```

```python
distances_df = pd.read_csv(data_dir + "distances.csv")
print(distances_df.columns)

N_1_accepted = []
N_2_accepted = []

for i, row in distances_df.iterrows():
    d1 = float(row['d1'])
    d2 = float(row['d2'])
    d3 = float(row['d3'])
    d4 = float(row['d4'])
    d5 = float(row['d5'])
    d6 = float(row['d6'])

    if d1 < 100 and d2 < 10 and d3 > 1e4:
        N_1_accepted.append(True)
    
    else:
        N_1_accepted.append(False)
        
    
    if d4 < 100 and d5 < 10 and d6 > 1e4:
        N_2_accepted.append(True)
    
    else:
        N_2_accepted.append(False)
    

distances_df['N_1_accepted'] = N_1_accepted
distances_df['N_2_accepted'] = N_2_accepted

    
```

```python
# Get file paths
file_paths = [file_path for file_path in glob.iglob(states_dir + "*.csv")]

# Reorder paths
ordered_paths = [0 for f in file_paths]
print(np.shape(ordered_paths))
for f in file_paths:
    file_name = os.path.basename(f)
    split_name = file_name.split('_')
    sim_idx = int(split_name[5])
    ordered_paths[sim_idx] = f
```

```python
style.use('seaborn-muted')

particle_accepted = distances_df['Accepted'].values

fig, ax = plt.subplots(nrows=3)
ax_num = 0
sns.set_style("white")
sns.set_context("talk")

pal = sns.color_palette()

for idx, file_path in enumerate(ordered_paths):
    sim_data = np.genfromtxt(file_path, dtype=float, delimiter=',')
    sim_shape = np.shape(sim_data)
    
    if len(sim_data) <= 7:
        continue
        
    if particle_accepted[idx]:
        if idx == 1 or idx ==8:
            continue
        t = np.arange(0, sim_shape[0], step=1)
        sns.lineplot(t, sim_data[:, 0], ax=ax[ax_num], alpha=1, color=pal[0])
        ax_num += 1


for i in ax:
    i.set_yscale('log')
    i.spines["right"].set_visible(False)
    i.spines["top"].set_visible(False)
#     ax.spines["left"].set_visible(False)
#     ax.spines["bottom"].set_visible(False)
    i.spines["left"].set_alpha(0.5)
    i.spines["bottom"].set_alpha(0.5)


    i.set(yticks=[])
    i.set(xticks=[])

    i.set_xlabel('Time')
    i.set_ylabel('Population')

plt.savefig(output_dir + "accepted_sims.pdf", dpi=500)
```

```python

N_1_accepted = distances_df['N_1_accepted'].values
N_2_accepted = distances_df['N_2_accepted'].values


fig, ax = plt.subplots(nrows=3)

ax_num = 0

sns.set_style("white")
sns.set_context("poster")
pal = sns.color_palette()

for idx, file_path in enumerate(ordered_paths):
    sim_data = np.genfromtxt(file_path, dtype=float, delimiter=',')
    
    sim_shape = np.shape(sim_data)
    

    if len(sim_data) <= 7:
        continue
    
    has_negative = (sim_data<0).any(axis=0)

    if not particle_accepted[idx] and sim_shape[0] ==1000:
        t = np.arange(0, sim_shape[0], step=1)
        
        if not N_1_accepted[idx]:
            if idx == 3:
                continue
            sns.lineplot(t, sim_data[:, 0], ax=ax[ax_num], alpha=1, color=pal[2])
            ax_num +=1
            
#         if not N_2_accepted[idx]:
#             sns.lineplot(t, sim_data[:, 1], ax=ax, alpha=0.5)


for i in ax:
    i.set_yscale('log')
    i.spines["right"].set_visible(False)
    i.spines["top"].set_visible(False)
    
    i.spines["left"].set_alpha(0.5)
    i.spines["bottom"].set_alpha(0.5)

    # ax.spines["left"].set_visible(False)
#     i.spines["bottom"].set_visible(False)

    i.set(yticks=[])
    i.set(xticks=[])

    i.set_ylabel('Population')
    i.set_xlabel('Time')

plt.savefig(output_dir + "rejected_sims.pdf", dpi=500)

```

```python
nrows = 2
ncols = 3
fig, ax = plt.subplots(nrows=nrows, ncols=ncols)

for row in range(nrows):
    for col in range(ncols):
        lwr = np.random.uniform(0, 0.5)
        upr = np.random.uniform(lwr, 1.0)
        data = upr/2 + lwr

        bins = [lwr, upr]
        sns.distplot(data, bins, ax=ax[row, col], kde=False, color=pal[0])

        
greek_letters = [r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$',r'$\epsilon$', r'$\zeta$']

count = 0
for i in ax:
    for j in i:
        j.spines["right"].set_visible(False)
        j.spines["top"].set_visible(False)

        j.spines["left"].set_alpha(0.5)
        j.spines["bottom"].set_alpha(0.5)
        
        j.set_xlim(0, 1)
        j.set(yticks=[])
        j.set(xticks=[])
        
        param = greek_letters[count]
        ylabel = 'p(' + param + ')|M)'
        j.set_ylabel(ylabel)
        
        count += 1
        
```
